```python
sns.boxplot(data = df,y="Beer_PerCapita",x="Hemisphere")
plt.title('Beer Consumption PerCapita in differnt Hemisphere')
plt.savefig(".Features of Species.png", dpi=1000,bbox_inches='tight') #dpi is the image size, have to save the fig before showing the plot!!
plt.show()
```

```python
Total_alcohol_consumption = df["Beer_PerCapita"] + df["Spirit_PerCapita"]+ df["Wine_PerCapita"]
df["Total_alcohol_consumption_PerCapita"] = Total_alcohol_consumption
print(df)
```

```python
df.groupby("Hemisphere").Total_alcohol_consumption_PerCapita.mean().plot(kind='bar') 
```

```python
df[["Spirit_PerCapita","Country", "Region"]].sort_values(by='Spirit_PerCapita', ascending=False )
```

```python
sns.boxplot(data = df,y="Total_alcohol_consumption_PerCapita",x="Region")
plt.title('Alcohol consumption in differnt Region')
plt.savefig(".Regionalconsumption.png", dpi=1000,bbox_inches='tight') #dpi is the image size, have to save the fig before showing the plot!!
plt.show()
```
