# Plotting the data using Pandas/Matplotlib

```python
iris.species.value_counts()  # Pandas series
```

```python
# Pandas bar plot
# Number of rows per species 
iris.species.value_counts().plot(kind='bar', figsize=(9,6)) 
plt.title('Data Balance')
plt.xlabel('Species')
plt.xticks(rotation=45) #rotating the x-axis labels by 45 degrees
plt.show()
```

```python
# change colors 
iris.species.\
value_counts().plot(kind='bar', figsize=(6,4),
                    color=['red','green',
                           'blue','pink']) 
```

```python
# Add plot information, attributes can be called with a variable name
colors = ['#E8DAEF','purple','xkcd:pastel blue'] #different ways to source colors
iris.species.value_counts().plot(kind='bar', figsize=(9,6), color = colors) 
plt.xlabel('Species')
plt.xticks(rotation=45)
plt.ylabel('Count of Species')
plt.title('Iris : Count of Data Points by Species')
#plt.savefig("./Iris Plots/Species Count.png", dpi=720,bbox_inches='tight')
plt.show()
```

```python
# check the mean of features by species
iris.groupby('species').mean()
```

Petal width looks like a good feature to classify setosa due to its low mean compared to the others.

```python
colors = ['#E8DAEF','purple','xkcd:pastel blue','pink']
iris.groupby('species').mean().plot(kind='bar', figsize=(9,6), color = colors) 
plt.title('Iris : Sepal & Petal Features by Species')
#plt.savefig("./Iris Plots/Features of Species.png", dpi=720,bbox_inches='tight')
plt.show()
```

kde plot - kernel density estimate

```python
plt.style.use('seaborn-pastel')
```

Let's take a closer look by using `.kde` to visualize the distributions.

```python
iris.plot.kde(figsize=(9,6))
```

Looking at the dataset, we can see that two of the features: petal width and petal length are bimodal, this gives us some intution that they may potentially be good features to separate the classes by.

We can group by species to take a closer look at petal width. We clearly see that petal width would be a good feature to separate the classes of flowers on, especially for setosa since the distributions don't overlap much, if at all.

```python
iris.groupby('species').petal_width.plot.kde(figsize=(6,4))
plt.legend(loc='center right', fontsize=10)
```

In contrast, sepal length isn't too great of a feature to separate the flowers on, due to the greater overlap in the distributions as seen below.

```python
s1=iris.groupby('species').sepal_length.plot.kde(figsize=(7,5))
plt.legend(loc='upper right', fontsize=10)
```
