## Web Scrap/ Crawl:

```python
# import BeautifulSoup library
from bs4 import BeautifulSoup
import requests

#Usage for BS:
#- parsing HTML (Data displayed driven) and XML (Data stroage driven) documents <> header tags will tell you xml/html
    #- XML is abbreviation for extensible Markup Language whereas HTML stands for Hypertext Markup Language.
    # XML mainly focuses on transfer of data while HTML is focused on presentation of the data.
    # XML is content driven whereas HTML is format driven.
    # XML is Case sensitive while HTML is Case insensitive.
    # XML provides namespaces support while HTML doesnâ€™t provide namespaces support.
    # XML tags are extensible whereas HTML has limited tags.
    # XML tags are not predefined whereas HTML has predefined tags. 

#Documentation: https://beautiful-soup-4.readthedocs.io/en/latest/

#In depth - https://www.dataquest.io/blog/web-scraping-beautifulsoup/
```

## Beautiful soup demo - IMDB

```python
from requests import get
url = 'https://www.imdb.com/search/title?release_date=2017&sort=num_votes,desc&page=1'
URL="http://quotes.toscrape.com"
response = get(URL)
#type(response)
print(response.text[:1000]) #After understanding the structuring of html 
```

```python
# CLick on the div drop down menu - for more in depth analysis of IMDB webpage (Inspect mode)

from bs4 import BeautifulSoup
html_soup = BeautifulSoup(response.text, 'html.parser')
type(html_soup)

print(html_soup.prettify())
```

```python
movie_containers = html_soup.find_all('div', class_ = 'lister-item mode-advanced')
print(type(movie_containers)) #Result set object, it will contain everything we need - Movie name, IMDB score, rating, meta scores...
print(movie_containers)

#print(len(movie_containers))
```
