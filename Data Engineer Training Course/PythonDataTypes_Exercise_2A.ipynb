6. Return the total number of lists inside a given list.

```
eg.
for [[1, 2, 3], [1, 2, 3], [1, 2, 3], [1, 2, 3]] there are 4 sub-lists inside the list
```

````python
my_list = [[1, 2, 3], [1, 2, 3], [1, 2, 3], [1, 2, 3]]
sublist_count = 0

for item in my_list:
    if isinstance(item, list):
        sublist_count += 1

print(sublist_count)
````

9. Take an integer and output an n x n square solely consisting of the integer n.

```
eg.
For n = 3,
the output should be

3, 3, 3
3, 3, 3
3, 3, 3
```

````python
n = 3

square = [[n for _ in range(n)] for _ in range(n)]

for row in square:
    print(*row)
````

10. Given a phrase, return False if any individual word in the phrase contains duplicate letters. Return True otherwise.

````python
def check_duplicate_letters(phrase):
    words = phrase.split()

    for word in words:
        if len(set(word)) != len(word):
            return False

    return True

# Example usage
phrase = "hello world"
result = check_duplicate_letters(phrase)
print(result)
````
