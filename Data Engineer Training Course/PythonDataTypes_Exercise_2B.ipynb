9. Given the string "It is a far, far better thing that I do, than I have ever done; it is a far, far better rest that I go to than I have ever known.", create a new string "thing rest" from it.

````python
original_string = "It is a far, far better thing that I do, than I have ever done; it is a far, far better rest that I go to than I have ever known."

# Split the original string into words
words = original_string.split()

# Extract the required words "thing" and "rest"
required_words = [word for word in words if word.lower() in ['thing', 'rest']]

# Create the new string by joining the required words
new_string = ' '.join(required_words)

print(new_string)
````

10. Find the list in a list of lists whose sum of elements is the highest.

```
eg.
for the list [[2,3,4], [4,5,6], [10,11,12], [7,8,9]]
the sublist [10,11,12] has the greatest sum of elements
```

````python
list_of_lists = [[2, 3, 4], [4, 5, 6], [10, 11, 12], [7, 8, 9]]

max_sum = float('-inf')  # Initialize max_sum with negative infinity
max_sublist = None

for sublist in list_of_lists:
    sublist_sum = sum(sublist)
    if sublist_sum > max_sum:
        max_sum = sublist_sum
        max_sublist = sublist

print("The sublist",max_sublist,"has highest sum of elements")
````
