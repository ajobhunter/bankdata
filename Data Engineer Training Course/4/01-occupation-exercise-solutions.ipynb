**5)** What is the male to female ratio for each occupation?



```python
d = {'Occupation': [], 'M:F Ratio': []}

for i in df.groupby('occupation').groups:
    a = df.groupby('occupation').get_group(i)['gender'].value_counts()
    if 'F' not in a:
        d['Occupation'].append(i)
        d['M:F Ratio'].append('inf')
        continue
    d['Occupation'].append(i)
    d['M:F Ratio'].append(a['M']/a['F'])

ratio_df = pd.DataFrame(data=d)
#d
ratio_df
```
