## 4. Positional Arguments

Let’s return to Danger's Restaurant. The greeting of the day will not always be the same. In this case. we can use pass arguments into the function when we call it:

````python
# Example of positional argument

def welcome_message(greeting):
    print(f"{greeting}! welcome to Danger Restaurant!")

welcome_message("Good Morning")
welcome_message("Good Afternoon")
````

In the definition for `welcome_message()`, the  `greeting` is referred to as formal parameter. This variable name is a placeholder for the name of the greeting message. whenever we call `welcome_message`, we have to provide `greeting`:

The value inside the parentheses is referred to as an **argument** of the function. The **argument** is the information that is to be used in the execution of the function. When we call the function, the formal parameter `greeting` is assigned with the actual parameter data "Good morning"

Every time we call `welcome_message()` with a different value inside the parentheses, `greeting` is assigned to hold that value

We can make a function take more than one parameter by using commas:

````python
def welcome_message(
        restaurant_name,
        description_1,
        description_2):

    print(f"Good morning, Welcome to {restaurant_name}")
    print(description_1)
    print(description_2)

welcome_message(
    "Good morning",
    "We have breakfast",
    "We have drinks"
)
````

## 6. Keyword Arguments

In our `welcome_message()` function, we had 4 arguments:

````python
def welcome_message(
        restaurant_name,
        description_1,
        description_2,
        description_3):

    print(f"Good morning, Welcome to {restaurant_name}")
    print(description_1)
    print(description_2)
    print(description_3)
````

Whichever value is put into `welcome_message()` first is assigned to `restaurant_name`, and whichever value is put in second is assigned to `breakfast_description_2`, etc. These are called positional arguments because their assignments depend on their positions in the function call.

We can also pass these arguments as keyword arguments, where we explicitly refer to what each argument is assigned to in the function call.

````python
welcome_message(
    description_1='Cheese Burger',
    description_2='Company Sandwiches',
    restaurant_name='Danger\'s Restaurant',
    description_3='Croissant'
)
````

We can also define default arguments for a function. If the function is called without an argument for that parameter, it relies on the default.

````python
def welcome_message(
        restaurant_name,
        description_1="Cheese Burger",
        description_2="Sandwiches",
        description_3="Croissant"):
    print(f"Good morning, Welcome to {restaurant_name}")
    print(description_1)
    print(description_2)
    print(description_3)
````

In this case, all breakfast_description have default value. If we call the function with only one argument, the value is used for `restaurant_name`:

````python
# Default value will be used if optional args are not specified
welcome_message('Danger\'s Restaurant')

# We can also override the default values when specified
welcome_message(
    'Danger\'s Restaurant',
    description_1='Hehe',
    description_2='Haha')
````

## 7. Mixing positional arguments and keyword arguments

````python
def print_something(item1, item2="10", item3="50"):
    print(f"Item 1: {item1}")
    print(f"Item 2: {item2}")
    print(f"Item 3: {item3}")
````

````python
# Try it
print_something(1,1,10)
# print_something(1, 2, 3)
# print_something(1)
# print_something(item3='10')
# print_something(1, item3='10', item2='1')
````

**Rules:**

1. Positional arguments always go before keyword arguments
2. All position arguments must be provided, but kwargs are optional
3. Positional arguments is turned into kwargs automatically according to the order in function defination

## Unpacking Arguments (Optional)

`list`s and `dict`s can be unpacked as positional arguments and keyword arguments respectively.

With unpacking, you can handle unknown number of arguments or new keyword arguments

**Note**: The rule that positional arguments precedes keyword arguments still holds

````python
# Example

def unpack_args(*args):
    print(f"You have entered {len(args)} positional arguments")
    for i in range(len(args)):
        print(f"args[{i}] = {args[i]}")

def unpack_kwargs(**kwargs):
    print(f"You have entered {len(kwargs)} keyword arguemnts")
    for keyword, argument in kwargs.items():
        print(f"kwargs['{keyword}'] = {argument}")

unpack_args("a", "b", "c")
print("="*10)
unpack_kwargs(a=1, b=2, c=3)
````

## 8. Returns

Functions can also return a value to the user so that this value can be modified or used later.

Here is an example of a function `calculate_age` that sums two arguments and returns the result:

````python
def calculate_age(current_year, birth_year):
  age = current_year - birth_year
  return age

my_age = calculate_age(2020, 2002)
print(my_age)
````

The returned result can be re-used:

````python
print(f"My age is {my_age}")
````

## 10. Multiple Return Values

Sometimes we may want to return more than one value from a function. We can return several values by separating them with a comma.


````python
def square_point(x_value, y_value):
    x_2 = x_value * x_value
    y_2 = y_value * y_value
    return x_2, y_2
````

This function takes in an x value and a y value, and returns both squared values. We can get those values by assigning them both to variables when we call the function:

````python
x_squared_value, y_squared_value = square_point(5, 7)
print(x_squared_value)
print(y_squared_value)
````
