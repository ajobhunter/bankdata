Import libraries

```python
# Import necessary libraries
# pandas, numpy, matplotlib, seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load data

```python
# Load the dataset from a CSV file with specified encoding
#https://www.kaggle.com/datasets/nelgiriyewithana/countries-of-the-world-2023


df = pd.read_csv("world-data-2023.csv", encoding="ISO-8859-1")
```

Data Cleaning and Preparation

```python
df.columns
```

```python
#select some column
df = df[['Country','Birth Rate','Fertility Rate','Infant mortality','Life expectancy','Physicians per thousand','Population']]
```

```python
df.info()
```

```python
df
```

**The following columns have missing values:**

Birth Rate

Fertility Rate

Infant mortality

Life expectancy

Physicians per thousand

Population (currently of object type due to the presence of commas in the numbers).

```python
# Convert 'Population' to strings before replacing commas 使用 `astype( 將 'Population' 列轉換為字串str)`方法。 然後，它使用“str.replace()”方法刪除人口值字串表示形式中的所有逗號。
df['Population'] = df['Population'].astype(str).str.replace(',', '')

# Convert 'Population' to numeric with 'coerce' option 使用 `pd.to_numeric()` 將 'Population' 欄位轉換為數值功能。 `errors='coerce'` 選項用於將任何無效或非數字值替換為 NaN（缺失值）。
df['Population'] = pd.to_numeric(df['Population'], errors='coerce')

# Columns to impute 創建一個名為`columns_to_impute` 的列表，其中包含需要估算（即填充）缺失值的列的名稱。
columns_to_impute = ['Birth Rate', 'Fertility Rate', 'Infant mortality', 'Life expectancy', 'Physicians per thousand', 'Population']

# Impute missing values with column medians
median_values = df[columns_to_impute].median(numeric_only=True) #使用 `median()` 方法計算 `columns_to_impute` 中指定的列的中位數。 “numeric_only=True”參數確保僅考慮數字列來計算中位數。
df[columns_to_impute] = df[columns_to_impute].fillna(median_values) #使用 `fillna()` 方法以對應的中位數填入 `columns_to_impute` 中指定的列中的缺失值。
```

```python
# Rename columns for consistency
df.rename(columns={'Birth Rate': 'Birth_Rate'},inplace = True)
df.rename(columns={'Fertility Rate': 'Fertility_Rate'},inplace = True)
df.rename(columns={'Infant mortality': 'Infant_Mortality'},inplace = True)
df.rename(columns={'Life expectancy': 'Life_Expectancy'},inplace = True)
df.rename(columns={'Physicians per thousand': 'Physicians_Per_Thousand'},inplace = True)
df.columns
```

**This demonstrates some intresting and strong correlations between features of the dataset.**

```python
# let's explore some of this further
# Calculate and display correlation matrix heatmap

correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
```

```python
# Display distribution histograms for each column
plt.figure(figsize=(12, 8))

for column in df.columns[1:]:
    plt.subplot(2, 3, df.columns.get_loc(column))
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
```
