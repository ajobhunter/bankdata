```python
#web scraping weather
#https://www.youtube.com/watch?v=YsC5jc6OKFc
import requests
from bs4 import BeautifulSoup

url="https://www.timeanddate.com/weather/hong-kong/hong-kong"
req=requests.session().get(url)
soup=BeautifulSoup(req.text,features="html.parser")
result=soup.select(".h2")[0].text

print(result)
#self convert °F to °C
```
