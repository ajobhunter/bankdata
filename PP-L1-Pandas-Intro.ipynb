### 1. Create a DataFrame - by column

- A DataFrame is an object that stores data as rows and columns.

- Think of a DataFrame as a spreadsheet or as a SQL table.

- You can manually create a DataFrame or fill it with data from a CSV, an Excel spreadsheet, or a SQL query (more is coming).

- DataFrames have rows and columns. Each column has a name, which is a string. Each row has an index, which is (usually) an integer.

- You can pass in a dictionary to `pd.DataFrame()`. Each key is a column name and each value is a list of column values. The columns must all be the same length or you will get an error. Here’s an example:

````python
# Example of creating a DataFrame

df1 = pd.DataFrame({
    "name": ["John Smith", "Jane Doe", "Joe Schmo"],
    "address": ["123 Main St.", "456 Maple Ave.", "789 Broadway"],
    "age": [34, 28, 51]
})

df1
````

### 3. Comma Separated Values (CSV)

- Most of the time, we’ll be working with datasets that already exist.

- One of the most common formats for big datasets is the CSV.

- CSV (comma separated values) is a text-only spreadsheet format. The first row of a CSV contains column headings. All subsequent rows contain values. Each column heading and each variable is separated by a comma:

```
key1,key2,key3
value1,value2,value3
value4,value5,value6
```

### 4. Loading and Saving CSVs

When you have data in a CSV, you can load it into a DataFrame in Pandas using `.read_csv()`:
```python
df = pd.read_csv("my-csv-file.csv")
```
In the example above, the `.read_csv()` method is called. The CSV file called my-csv-file is passed in as an argument.

We can also save data to a CSV, using `.to_csv()`:
```python
df.to_csv('new-csv-file.csv')
```
In the example above, the `.to_csv()` method is called on df (which represents a DataFrame object). The name of the CSV file is passed in as an argument (new-csv-file.csv). By default, this method will save the CSV file in your current directory.

Remarks: Row index could be annoying sometimes, adding ``` index = False``` in ```to_csv()``` could cure your headache.

### 6. Select Columns

Suppose you have the DataFrame called customers, which contains the ages of your customers:

image.png

Perhaps you want to take the average of the ages. In order to do the task, you’d need to select the column.

There are two possible syntaxes for selecting all values from a column:

1. Select the column as if you were selecting a value from a dictionary using a key. In our example, we would type `customers['age']` to select the ages.<br>

2. If the name of a column follows all of the rules for a variable name (doesn’t start with a number, doesn’t contain spaces or special characters, etc.), then you can select it using the following notation: df.MyColumn. In our example, we would type `customers.age`.<br>

When we select a single column, the result is called a `Series`.

### 9. Selecting Multiple Rows

You can also select multiple rows from a DataFrame. Here are some different ways of selecting multiple rows (Just like what you do to a List!)

`df.iloc[3:7]` would select all rows starting at the 3rd row and up to but not including the 7th row (i.e., the 3rd row, 4th row, 5th row, and 6th row)

`df.iloc[:4]` would select all rows up to, but not including the 4th row (i.e., the 0th, 1st, 2nd, and 3rd rows)

`df.iloc[-3:]` would select the rows starting at the 3rd to last row and up to and including the final row

###  10. Select Rows with boolean series

You can select a subset of a DataFrame by using logical statements:
```python
df[df.MyColumnName == desired_column_value]
```

image.png

Suppose we want to select all rows where the customer’s age is 30. We would use:

```python
df[df.age == 30]
```

We can use other logical statements, such as Greater Than, `>`, Less Than, `<` and Not Equal, `!=`

### 11. Combination of boolean series

You can also combine multiple logical statements, as long as each statement is in parentheses.

For instance, suppose we wanted to select all rows where the customer’s age was under 30 or the customer’s name was “Martha Jones”

```python
df[(df.age < 30) | (df.name == 'Martha Jones')]
```

For pandas series we use,
- `|` for “or”
- `&` for “and”
- `~` for “not”

### 12. More boolean series creation

Suppose we want to select the rows where the customer’s name is either “Martha Jones”, “Rose Tyler” or “Amy Pond”.
