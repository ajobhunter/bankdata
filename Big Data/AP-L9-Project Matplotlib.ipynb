#### Excercise

In the covid-19 dataset, plot multiple lines and highlight the line for Hong Kong in red.  Here's demo:
image.png

```python
import pandas as pd
df_covid = pd.read_csv("https://drive.google.com/u/1/uc?id=16n8w-JtvlwJeX7Ft7uikvUqffEJrcwyl&export=download")

df_covid
```

```python
# clean data of column: province_state
df_clean = df_covid[~df_covid.province_state.isna()]
df_clean.head(20)
```

```python
# Check Hong Kong
df_clean[df_clean.province_state.str.startswith("Hong")]
```

```python
# Plot every line
# DDDA is a color code in RGBA format, A for alpha (transparency)
# DDD is a light grey, 9 is the degree of transparency
# .T is to transpose the table: flipping row and column
ax = df_clean.iloc[:, 5:].T.plot(legend=False, c="#DDD9")

# Plot Hong Kong
# F00 is red
# ax=ax tell matplotlib to plot on the same axe as above
df_clean[df_clean.province_state == "Hong Kong"].iloc[:, 5:].T.plot(legend=False, c="#F00", ax=ax)
```
