### Group by

Group by is an essential feature in every table manipulation. In pandas, we use the function
```python
# Group by a particular column
df.groupby("Column name")

# Group by some custom values
df.groupby(["A", "A", "B", "A", "B",.....])

```

The function returns a `pandas.core.groupby.generic.DataFrameGroupBy object` which is an intermediate object that stores grouping information.

We must give it further aggregation instruction so that we can obtain usable information.
```python
# Follow the groupby by a sum()
df.groupby("Column name").sum()

# Follow the groupby by a max()
df.groupby("Column name").max()

# Follow the groupby by your own function
df.groupby("Column name").agg(func)

# Follow the groupby by your multiple functions
df.groupby("Column name").agg([func1, func2])

# Follow the groupby by different functions for different columns
df.groupby("Column name").agg({"col1": func1, "col2": func2})
```

**Advanced group by**

Sometimes we need more granular control towards group by.

E.g.
1. Columns is a multi-index
2. Column values need to be binned before group


```python

# The normal
df.groupby(pd.Grouper(key="column name"))

# To group a date column with binning
df.groupby(pd.Grouper(key="column name", freq="3m"))

```

**Binning**

Very often we need to put values into bins.  

While the `pd.Grouper(freq="")` functionality is designed for time, we have `pd.cut()` to create bins for regular numbers.

Each bin will have equal length.

```python
df.groupby(pd.cut(series, bins=100)).count()
```

*Histogram*: Binning and counting is essentially the basic for creating a histogram.

Pandas provide an easier method for that

```python
series.hist(bins=100)
```

**Group by for Series**

For series, we do not have separated columns as the group by column and the value columns, so we need extra values to group by.

```python
series.groupby(["A", "A", "B", "A", "B",.....])
```

**Short cut**

We have an equivalent for `series.groupby(series).count()`


```python
series.value_counts()
```
