````python
# %load solutions/count-1.py
def count_chars(text):
    
    l = list(text) #convert text into list
    d = {} #d=dict empty dict to store character counts
    for char in l:
        if char not in d:
            d[char] = 1
        else:
            d[char] +=1
    return d
    

def count_words(text):
    
    l = text.split()
    d = {}
    for word in l:
        if word not in d:
            d[word] = 1
        else:
            d[word] +=1
            
    return d
    

word_freqs = count_words(text)
n_words = sum(word_freqs.values())
print("Number of Words: ",n_words)

#calculates the number of sentences in the text variable 
#by using the count_chars function to count the occurrences of periods ('.').
char_freqs = count_chars(text)
n_sents = char_freqs['.']
print("Number of Sentences: ",n_sents)
````
