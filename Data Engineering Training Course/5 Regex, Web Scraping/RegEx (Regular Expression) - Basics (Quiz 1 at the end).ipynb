## Q1: Important note: r"" ? ""? Which string should I use?

### Answer: in short, use r"" (Convention), another interesting fact

### Without raw strings, we could missout important information from our data


```python
string0 = "Hello\nWorld" #\n = new line 
print(string0, end= '\n\n')

string0 = r"Hello\nWorld" #r means raw
print(string0, end= "\n\n")
```

```python
string1 = "Howdy, partner, Paw."
```

## Negated/ Excluded Sets

```python

re.findall("[^aeiou]", string1)
```

# Dots

```python
string3 = "Hello World ... \
how are you doing"
print(string3)
```

```python
print("The number of dots inside of this string is = ", len(re.findall(r"[.]", string3))) # Recovering periods/ dots
```

```python
re.findall(r".", string3) # Returning everything except line breaks 
```

## Match Any:  \S\s

```python
re.findall(r"[\s\S]", string3) # Returning everything except line breaks (also the same as r"[\w\W]")
```

## Whitespaces/ Non-whitespaces: \s \S (Meta characters/ Escaped Chars)

```python
re.findall(r"[\S]", string3) # Returning everything but line breaks/ spaces
```

```python
print(string3)
re.findall(r"[\s]", string3) #Returns only whitespaces

len(re.findall(r"[\s]", string3))
```

## Be careful and don't forget your char set brackets

```python
# What we establiashed already

re.findall(r"[\s\S]", string3) # Returning whitespaces + non-whitespaces (that's everything, all chars)
```

```python
print(string3)

re.findall(r"\s\S", string3) # Q4: ?? What's going on here? 
```

```python
re.findall(r"\S\s", string3) # Ditto, what's going on here?
```

## Word and non-word \w \W

```python
re.findall(r"\W", string3) #Non-word # Same as saying [^A-Za-z0-9_]
```

```python
re.findall(r"\w", string3)#Non-word # Same as saying [A-Za-z0-9_]
```

## Digit and non-digit \d \D

```python
string4 = "Holiday420"

re.findall(r"[\d]", string4)
```

```python
re.findall(r"[\D]", string4)
```
